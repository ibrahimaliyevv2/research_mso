```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('global gpu and cpu - 4.4.csv')

# Separate data for CPU and GPU
cpu_data = data[data['Type'] == 'Global CPU']
gpu_data = data[data['Type'] == 'Global GPU']

```

```python
# Create a box plot to compare running time (Time) for CPU and GPU
plt.figure(figsize=(8, 6))
plt.boxplot([cpu_data['Time'], gpu_data['Time']], labels=['Global CPU', 'Global GPU'])
plt.ylabel('Running Time (ms)')
plt.title('Comparison of Running Time (ms) between CPU and GPU')
plt.show()
```

```python
# Create a scatter plot to compare minimum values (Minimum) for CPU and GPU
plt.figure(figsize=(8, 6))
plt.scatter(cpu_data['Type'], cpu_data['Minimum'], color='blue', label='Global CPU')
plt.scatter(gpu_data['Type'], gpu_data['Minimum'], color='green', label='Global GPU')
plt.ylabel('Minimum Value')
plt.title('Comparison of Minimum Values between CPU and GPU')
plt.legend()
plt.show()
```
